Filled missing values with the mode of the respective columns.

```python
most_common_artist = spotify_data['artist_name'].mode()[0]
spotify_data['artist_name'] = spotify_data['artist_name'].fillna(most_common_artist)
```

```python
most_common_genre = spotify_data['artist_genres'].mode()[0]
spotify_data['artist_genres'] = spotify_data['artist_genres'].fillna(most_common_genre)
```

Check for missing values in the dataset

```python
spotify_data.isnull().sum()
```

### Additional Features to Dataset

**Track Popularity Tier**
<br>Based on the 'track popularity' column, I created a new column named 'track_popularity_tier' to categorize tracks into different popularity tiers. </br>

```python
bins = [0, 20, 40, 60, 80, 100]
labels = ['Very Low', 'Low', 'Medium', 'High', 'Viral']
spotify_data['track_popularity_tier'] = pd.cut(spotify_data['track_popularity'], bins=bins, labels=labels, include_lowest=True)
spotify_data.head()
```

```python
spotify_data.tail()
```

**Artist Popularity Gap**
<br>Measures whether a track is over-performing or under-performing relative to the artist's overall popularity. </br>

```python
spotify_data['popularity_gap'] = spotify_data['track_popularity'] - spotify_data['artist_popularity']
spotify_data.head()
```

**Popularity Gap Type**
<br>Categorizes the 'popularity_gap' into "Breakout Track" and "Underperforming Track" </br>

```python
spotify_data['popularity_gap_type'] = spotify_data['popularity_gap'].apply(
    lambda x: 'Breakout Track' if x > 0 else 'Underperforming Track'
)
spotify_data.head()
```

```python
spotify_data.tail()
```

Updated the dataset with the new features and saved it as a new CSV file.

```python
spotify_data.to_csv("updated spotify_data clean.csv", index=False)
```

### Mutation/Transformation & Filtering

**Spotify Dataset Genres**
<br> Created a copy of the Spotify dataset and split rows so each track has one genre per row, then cleaned up extra whitespace in the genre names for analysis </br>

```python
spotify_genre = spotify_data.copy()
spotify_genre['artist_genres'] = spotify_genre['artist_genres'].str.split(",")
spotify_genre = spotify_genre.explode('artist_genres')
spotify_genre['artist_genres'] = spotify_genre['artist_genres'].str.strip()
spotify_genre.head()
```

Find the top 5 Spotify genres

```python
top_genres = spotify_genre['artist_genres'].value_counts().head(5).index

spotify_genre = spotify_genre[spotify_genre['artist_genres'].isin(top_genres)]
spotify_genre.head()
```

```python
genre_order = (
    spotify_genre.groupby('artist_genres')['track_popularity']
    .median()
    .sort_values()
    .index
)

genre_order
```

Find the 10 top artists based on the average of their track popularity

```python
top_artists = (
    spotify_data.groupby('artist_name')['track_popularity']
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

top_artists
```

Create an "artist_summary" table that displays the following:
- Average Track Popularity
- Number of Tracks
- Artist Popularity

```python
artist_summary = (
    spotify_data.groupby('artist_name')
    .agg(
        avg_track_popularity= ('track_popularity', "mean"),
        track_count = ("track_id", "count"),
        artist_popularity = ("artist_popularity", "mean")
    )
    .reset_index()
)

artist_summary
```

Add a new column to the "artist_summary" table that categorizes output levels of said artist based on how many tracks they have released. 

```python
artist_summary['output_group'] = pd.cut(artist_summary['track_count'],
                                        bins = [0, 10, 50, 200, 1000],
                                        labels = ['Low', 'Medium', 'High', 'Very High']
)

artist_summary
```

Select the 20 tracks with the largest popularity differences (positive or negative) and sort them so underperforming tracks appear below zero and breakout tracks above.

```python
top_popularity_gap = (
    spotify_data
    .assign(abs_gap=lambda x: x["popularity_gap"].abs())
    .sort_values("abs_gap", ascending=False)
    .head(20)
)

top_popularity_gap = top_popularity_gap.sort_values("popularity_gap")
top_popularity_gap
```

## 2: Data Visualizations Creation

**Histogram Plot**: Track Popularity Distribution by Explicit Content

<br> Chart Type Explanation </br>
<br> This chart was designed to explore whether explicit content is associated with different popularity patterns on Spotify. Specifically, it examines whether explicit and non-explicit tends to cluster at different popularity levels or show different distribution shapes. In order to achieve this exploration, I choose a histogram. This chart type is suitable for exploring the overall distribution of a numeric variable. By overlaying explicit and non-explicit tracks, the visualization allows direct comparison. </br>

<br> Steps to Create Visualization </br>
- Used sns.histplot() on the track_popularity column
- Added explicit as a hue to split the distribution into two group (explicit vs. non-explicit)
- Selected an appropriate number of bins to smooth the distribution (had to play around with this number for a while)
- Labeled axes and added a descriptive title

<br>Design Choices </br>
- Overlapping bars make it easy to compare distribution of explicit status
- Density normalization helps compare shapes rather than raw counts from the data

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")

spotify_palette = {
    False: "#1DB954",  # Spotify green
    True: "#191414"    # Black
}

sns.histplot(data = spotify_data, x= 'track_popularity', hue = 'explicit', bins = 20, stat = 'density', palette = spotify_palette, alpha = 0.55)
plt.xlabel('Track Popularity')
plt.ylabel('Density')
plt.title('Distribution of Track Popularity by Explicit Status')
plt.show()
```

**Preprocessing for Heatmap Visualization**
<br>Converted the 'popularity_tier' column to a categorical data type and created a heatmap to visualize the distribution of track popularity across different genres and years. </br>

```python
artist_bins = [0, 25, 50, 75, 100]
artist_labels = ['Low', 'Mid-Low', 'Mid-High', 'High']

spotify_data['artist_popularity_bin'] = pd.cut(spotify_data['artist_popularity'], bins=artist_bins, labels=artist_labels, include_lowest=True)
```

```python
heatmap_data = (
    spotify_data.groupby(['artist_popularity_bin', 'track_popularity_tier'])
    .size()
    .reset_index(name='count')
)

heatmap_data
```

```python
heatmap_pivot = heatmap_data.pivot(index='artist_popularity_bin', columns='track_popularity_tier', values='count')
heatmap_pivot
```

**Heatmap**: Track Popularity vs. Artist Popularity 

<br> Chart Type Explanation </br>
<br> This heatmap was used to illustrate the relationship between an artist's overall popularity and the popularity tier of their individual tracks. The goal was to understand whether highly popular artists consistently produce highly popular tracks. Thus, a heatmap was chosen as the most effective chart type since its great analyzing distributions of two categorical variables. By encoding frequency as color intensity, the visualization makes it easy to identify where tracks are most concentrated across combinations of artist popularity tiers and track popularity tiers.  

<br> Steps to Create Visualization </br>
- Preprocessing: Binned "artist_popularity" into 5 tiers using pd.cut()
- Used "track_popularity_tier" I created in the data mutation phase
- Grouped data by both tiers and counted frequencies
- Pivoted the table to a matrix format
- Plotted using sns.heatmap() with annotations

<br> Design Choices </br>
- The "YlGn" color palette emphasizes higher counters with darker shades and correlates to the Spotify color palette
- Cell annotations provide exact values for clarity
- Logical ordering of tiers helping the viewer read trends

```python
sns.heatmap(heatmap_pivot, annot=True, fmt="d", cmap="YlGn")
plt.title("Artist Popularity vs. Track Popularity Tier")   
plt.xlabel("Track Popularity Tier")
plt.ylabel("Artist Popularity Tier")
plt.tight_layout()
plt.show()
```
